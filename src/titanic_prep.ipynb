```python
## Import Block
from os import path, getcwd
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
import joblib
import json
```

```python
## Path to dataset
dataset = path.join(getcwd(), "..", "data", "titanic.csv")

## Read data
df = pd.read_csv(dataset)
df = df.dropna()

df.head(7)
```

```python
df = df.drop(columns=["PassengerId", "Name", "SibSp", "Ticket", "Embarked", "Cabin"], axis=1)
df.head(7)
```

```python
def resolve_gender(sex:str=None):
    female = 'female'
    male = 'male'
    if sex == female:
        return 0
    elif sex == male:
        return 1

def resolve_survival(survived=None):
    if survived == 1:
        return True
    elif survived == 0:
        return False
```

```python
df["Sex"] = df["Sex"].apply(resolve_gender)
df["Survived"] = df["Survived"].apply(resolve_survival).astype(str)
df.head(7)
```

```python
x = df.drop(columns=["Survived"])
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier(
    criterion="gini",
    splitter="best",
    max_depth=5,
    min_samples_leaf=5,
    min_samples_split=5,
    random_state=0
)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
joblib.dump(model, "titanic_survival_model.joblib")
```

```python
tree.export_graphviz(
    model,
    out_file="titanic_survival_tree.dot",
    feature_names=x.columns,
    class_names = sorted(y.unique()),
    label="all",
    rounded=True,
    filled=True
)
```
